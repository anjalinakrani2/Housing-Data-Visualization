```python
import numpy as np
import pandas as pd
```

```python
df=pd.read_csv("housing.csv")
```

```python
df.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

## **Boxplot for Median value of owner-occupied homes**

```python
plt.figure(figsize=(10,8))
sns.boxplot(x='MEDV',data=df)
plt.title('Median Value of Owner-Occupied Homes')
plt.show()
```

## **Histogram for the Charles River**

```python
plt.hist(df['CHAS'],bins=5,edgecolor='black')
plt.show()
```

```python
sns.catplot(x='CHAS',data=df,kind='count')
plt.title('Histogram for the Charles River')
plt.xlabel('Charles River')
plt.show()
```

## **Boxplot of MEDV vs Discretized AGE**

```python
df['AgeGroup']=pd.cut(df['AGE'],bins=[0,35,50,df['AGE'].max()],labels=['<=35','35-50','>50'])
df['AgeGroup']
```

```python
sns.boxplot(x='AgeGroup',y='MEDV',data=df)
plt.title("Boxplot of MEDV vs Discretized AGE")
plt.show()
```

### **Scatter Plot of NOX vs INDUS**

Nitric oxide concentrations and the proportion of non-retail business acres per town

```python
plt.figure(figsize=(8,6))
sns.scatterplot(x='NOX',y='INDUS',data=df)
plt.title('Scatter Plot of NOX vs INDUS')
plt.xlabel('Nitric oxide concentrations')
plt.ylabel('proportion of non-retail business acres per town')
plt.show()
```

```python
#****CONCLUSION***
print('The lower proportion of non-retail business acres per town had low nitric oxide concentrations')
```

## **Histogram for the pupil to teacher ratio**

```python
plt.figure(figsize=(10,8))
plt.hist(df['PTRATIO'],bins=20,edgecolor='black')
plt.show()
```

```python
sns.catplot(x='PTRATIO', kind='count', data=df, aspect=5)
plt.title('Histogram of Pupil to Teacher Ratio')
plt.show()
```
